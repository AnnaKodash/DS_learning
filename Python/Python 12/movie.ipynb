```python
import os

def concat_users_files(path):
    csv_list = os.listdir(path)
    df = pd.DataFrame()
    for i in csv_list:
        name = str(path)+str(i)
        csv_ = pd.read_csv(name)
        df = pd.concat([df, csv_], ignore_index=True)
    return df
```
