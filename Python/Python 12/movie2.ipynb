```python
import pandas as pd

ratings_movies = pd.read_csv('data/ratings_movies.csv')
```

```python
ratings_movies.head()
```

```python
import re
```

```python
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg)
    if len(candidates) > 0:
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return round(int(year))
    else:
        return None
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
```

```python
ratings_movies
```

```python
ratings_movies['year_release'].count()
```

```python
mask = ratings_movies['year_release']==1999
ratings_movies[mask].sort_values(by=['rating']).loc[:, ['rating', 'title']]
```

```python
ratings_movies[mask].groupby(by='title')['rating'].mean().sort_values()
```

```python
mask = ratings_movies['year_release']==2010
ratings_movies[mask].groupby(by='genres')['rating'].mean().sort_values()
```

```python
ratings_movies.groupby(by='userId')['genres'].nunique().sort_values(ascending=False)
```

```python
ratings_movies.groupby(by='userId')['rating'].agg(['count', 'mean']).\
    sort_values(by=['count', 'mean'], ascending=[True, False])
```

```python
mask = ratings_movies['year_release']==2018

grouped = ratings_movies[mask].groupby(by='genres')['rating'].agg('mean', 'count').sort_values(ascending=False)
```
