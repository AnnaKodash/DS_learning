```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.txt", delimiter= '\s+', index_col=False, header=None)
train
```

```python
test = pd.read_csv("data/test.txt", delimiter= '\s+', index_col=False, header=None) 
test
```

```python
data = pd.concat([train,test], axis=0)
data
```

```python
labels_train = pd.read_csv("data/train_labels.txt", delimiter= '\s+', index_col=False, header=None) 
labels_test = pd.read_csv("data/test_labels.txt", delimiter= '\s+', index_col=False, header=None) 
labels = pd.concat([labels_train,labels_test], axis=0)
labels
```

```python
labels.nunique()
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
X_norm = scaler.fit_transform(data)

print(round(X_norm[0][0], 2))
```

```python
from sklearn.metrics import silhouette_score
from sklearn import cluster
import seaborn as sns 
from sklearn import metrics

# напишем функцию, как и при подсчете метода локтя
def get_silhouette(cluster_num, X):
    k_means =  cluster.KMeans(n_clusters=cluster_num, init='k-means++', n_init=10, random_state=42)
    k_means.fit(X)
# подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
    silhouette = silhouette_score(X, k_means.predict(X))
    c_h = metrics.calinski_harabasz_score(X, k_means.predict(X))
    d_b = metrics.davies_bouldin_score(X, k_means.predict(X))
    return silhouette, c_h, d_b

# создадим пустой словарь, ключами будут инерция и количество кластеров
silhouette_res = {"silhouette": [], "cluster": []}
c_h_res = {"c_h": [], "cluster": []}
d_b_res = {"d_b": [], "cluster": []}

# выберем нужные данные 
X = X_norm

for cluster_num in range(2, 10):
    silhouette_res["silhouette"].append(get_silhouette(cluster_num, X)[0])
    silhouette_res["cluster"].append(cluster_num)
    c_h_res["c_h"].append(get_silhouette(cluster_num, X)[1])
    c_h_res["cluster"].append(cluster_num)
    d_b_res["d_b"].append(get_silhouette(cluster_num, X)[2])
    d_b_res["cluster"].append(cluster_num)
    
```

```python
data = pd.read_csv('data/Country-data.csv')
data.head()
```

```python
data['life_expec'].max()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(data.drop('country', axis=1))
X[0][0]
```

```python
X = pd.DataFrame(X)
X
```

```python
import seaborn as sns

sns.heatmap(X.corr(), annot=True)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5, random_state=42)
X_pca = pca.fit_transform(X)
X_pca[0].sum()
```

```python
X_pca = pd.DataFrame(X_pca)
sns.heatmap(X_pca.corr(), annot=True)
```

```python
from sklearn.metrics import silhouette_score
from sklearn import cluster

# напишем функцию, как и при подсчете метода локтя
def get_silhouette(cluster_num, X):
    k_means =  cluster.KMeans(n_clusters=cluster_num, init='k-means++', n_init=10, random_state=1)
    k_means.fit(X)
# подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
    silhouette = silhouette_score(X, k_means.predict(X))
    return silhouette

# создадим пустой словарь, ключами будут инерция и количество кластеров
silhouette_res = {"silhouette": [], "cluster": []}

# выберем нужные данные 
X = X_pca

for cluster_num in range(2, 11):
    silhouette_res["silhouette"].append(get_silhouette(cluster_num, X))
    silhouette_res["cluster"].append(cluster_num)
    
# сохраним в датафрейм значение силуэта и количество кластеров
silhouette_df = pd.DataFrame(silhouette_res)

# установим стиль для визуализиции
sns.set_style("darkgrid")
# визуализируем зависимость значения инерции от количества кластеров
sns.lineplot(data=silhouette_df, x="cluster", y="silhouette", marker= "o")
```

```python
kmeans = cluster.KMeans(n_clusters=4, random_state=1, init='k-means++', n_init=10)
predictions = kmeans.fit_predict(X_pca)
```
