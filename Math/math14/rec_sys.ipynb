```python
data = pd.read_csv('data/shared_articles.csv')
data.head()
```

```python
articles_df = data[data['eventType']=='CONTENT SHARED']
articles_df
```

```python
interactions_df = pd.read_csv('data/users_interactions.csv')
interactions_df.head()
```

```python
interactions_df.personId = interactions_df.personId.astype(str)
interactions_df.contentId = interactions_df.contentId.astype(str)
articles_df.contentId = articles_df.contentId.astype(str)
```

```python
event_type = {
   'VIEW': 1.0,
   'LIKE': 2.0, 
   'BOOKMARK': 2.5, 
   'FOLLOW': 3.0,
   'COMMENT CREATED': 4.0,  
}
```

```python
interactions_df['event_weight'] = interactions_df['eventType'].apply(lambda x: event_type[x])
```

```python
interactions_df.head()
```

```python
interactions_df['event_weight'].mean().round(2)
```

```python
users = interactions_df.groupby('personId')['contentId'].nunique()
```

```python
users = users[users>=5]
```

```python
merged = pd.merge(users, interactions_df, how='left', on='personId')
merged.head(1)
```

```python
merged['last_timestamp'] = (
    merged.groupby(['personId', 'contentId_y'])['timestamp'].max())
```

```python
merged['eventType'].count()
```

```python
def smooth_user_preference(x):
    return math.log(1+x, 2)
```

```python
sum_weight = merged.groupby(['personId', 'contentId_y'], as_index=False)['event_weight'].sum()
```

```python
sum_weight['event_weight'] = sum_weight['event_weight'].apply(smooth_user_preference)
sum_weight
```

```python
last_event = merged.groupby(['personId', 'contentId_y'], as_index=False)['timestamp'].max()
```

```python
result = pd.merge(sum_weight, last_event, on=['personId','contentId_y'])
result
```

```python
result['timestamp'].mean()
```

```python
result
```

```python
train_data = result[result['timestamp']<1475519545]
test_data = result[result['timestamp']>=1475519545]
```

```python
train_data
```

```python
final_df = (
    train_data.reset_index()
    .groupby('personId')['contentId_y'].agg(lambda x: list(x))
    .reset_index()
    .rename(columns={'contentId_y': 'true_train'})
    .set_index('personId')
)

final_df['true_test'] = (
    test_data.reset_index()
    .groupby('personId')['contentId_y'].agg(lambda x: list(x))
)

final_df['true_test'] = [ [] if x is np.NaN else x for x in final_df['true_test'] ]
final_df.head()
```

```python
final_df.info()
```

формирование списка рекомендаций (10 самых популярных позиций)

```python
popular = (
    train_data
    .groupby('contentId_y')
    .event_weight.sum().reset_index()
    .sort_values(by='event_weight', ascending=False)
    .contentId_y.values
)
popular[0]
```

precision с отсечкой 10

```python
top_k = 10
 
final_df['popular'] = (
    final_df.true_train
    .apply(
        lambda x:
        popular[~np.in1d(popular, x)][:top_k]
    )
)
def calc_precision(column):
    return (
        final_df
        .apply(
            lambda row:
            len(set(row['true_test']).intersection(
                set(row[column]))) /
            min(len(row['true_test']) + 0.001, 10.0),
            axis=1)).mean()
calc_precision('popular')
```
