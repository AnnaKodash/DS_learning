```python
x, y, z = symbols('x y z')
```

```python
def f(x, y, z):
    return x**3-2*x**2+y**2+z**2-2*x*y+x*z-y*z+3*z

def grad(x, y, z):
    dx = 3*x**2 - 4*x - 2*y + z
    dy = -2*x + 2*y - z
    dz = x - y + 2*z + 3
    return (dx, dy, dz)
```

```python
func = x**3-2*x**2+y**2+z**2-2*x*y+x*z-y*z+3*z
print(func.diff(x))
print(func.diff(y))
print(func.diff(z))
```

```python
x0 = (1,1,1)
gamma = 0.01
x_cur = x0

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1],
            x_cur[2]-gamma*grad(*x_cur)[2])
    if i > 5000:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.4f}, {x_cur[1]:.4f}, {x_cur[2]:.4f});"
          f" f(x)={f(*x_cur):.4f}; grad f(x)=({grad(*x_cur)[0]:.4f}, {grad(*x_cur)[1]:.4f}, {grad(*x_cur)[2]:.4f})")
```

```python
from scipy import optimize
```

```python
optimize.minimize(lambda x: f(*x), x0=(1,1,1)) 
```
