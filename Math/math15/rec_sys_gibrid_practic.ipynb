```python
import pandas as pd
import math
import numpy as np
```

```python
data = pd.read_csv('shared_articles.csv')
data.head()
```

```python
articles_df = data[data['eventType']=='CONTENT SHARED']
articles_df
```

```python
interactions_df = pd.read_csv('users_interactions.csv')
interactions_df.head()
```

```python
interactions_df.personId = interactions_df.personId.astype(str)
interactions_df.contentId = interactions_df.contentId.astype(str)
articles_df.contentId = articles_df.contentId.astype(str)
```

```python
event_type = {
   'VIEW': 1.0,
   'LIKE': 2.0,
   'BOOKMARK': 2.5,
   'FOLLOW': 3.0,
   'COMMENT CREATED': 4.0,
}
```

```python
interactions_df['event_weight'] = interactions_df['eventType'].apply(lambda x: event_type[x])
```

```python
interactions_df.head()
```

```python
interactions_df['event_weight'].mean().round(2)
```

```python
users = interactions_df.groupby('personId')['contentId'].nunique()
```

```python
users = users[users>=5]
```

```python
merged = pd.merge(users, interactions_df, how='left', on='personId')
merged.head(1)
```

```python
merged['eventType'].count()
```

```python
def smooth_user_preference(x):
    return math.log(1+x, 2)
```

```python
sum_weight = merged.groupby(['personId', 'contentId_y'], as_index=False)['event_weight'].sum()
```

```python
sum_weight['event_weight'] = sum_weight['event_weight'].apply(smooth_user_preference)
sum_weight
```

```python
last_event = merged.groupby(['personId', 'contentId_y'], as_index=False)['timestamp'].max()
```

```python
result = pd.merge(sum_weight, last_event, on=['personId','contentId_y'])
result
```

```python
result['timestamp'].mean()
```

```python
result
```

```python
train_data = result[result['timestamp']<1475519545]
test_data = result[result['timestamp']>=1475519545]
```

```python
train_data
```

```python
from pandas.core.reshape.pivot import pivot_table
ratings_m = pivot_table(train_data, values='event_weight', columns='contentId_y', index='personId').fillna(0)
```

```python
ratings_m.loc['-1032019229384696495']['943818026930898372']
```

```python
ratings = np.asarray(ratings_m)
```

```python
round(ratings.mean(),3)
```

```python
similarity_users = np.zeros((len(ratings), len(ratings)))
for i in (range(len(ratings)-1)):
    for j in range(i+1, len(ratings)):

        mask_uv = (ratings[i] != 0) & (ratings[j] != 0)
        ratings_v = ratings[i, mask_uv]
        ratings_u = ratings[j, mask_uv]

        similarity_users[i,j] = np.corrcoef(ratings_v, ratings_u)[0, 1]
        similarity_users[j,i] = similarity_users[i,j]

similarity_users[3,40]
```

```python
similarity_users
```

```python
interactions = (
    train_data
    .groupby('personId')['contentId_y'].agg(lambda x: list(x))
    .reset_index()
    .rename(columns={'contentId_y': 'true_train'})
    .set_index('personId')
)

interactions['true_test'] = (
    test_data
    .groupby('personId')['contentId_y'].agg(lambda x: list(x))
)

interactions['true_test'] = [ [] if x is np.NaN else x for x in interactions['true_test'] ]

prediction_user_based = []
for i in range(len(similarity_users)):
    users_sim = similarity_users[i] > 0
    if not any(users_sim):
        prediction_user_based.append([])
    else:
        tmp_recommend = np.argsort(ratings[users_sim].sum(axis=0))[::-1]
        tmp_recommend = ratings_m.columns[tmp_recommend]
        recommend = np.array(tmp_recommend)[~np.in1d(tmp_recommend, interactions.iloc[i]["true_train"])][:10]
        prediction_user_based.append(list(recommend))
interactions['prediction_user_based'] = prediction_user_based
prediction_user_based[35][0]
```

```python
def calc_precision(column):
    return ( interactions.apply(  lambda row:len(set(row['true_test']).intersection(
                set(row[column]))) /min(len(row['true_test']) + 0.001, 10.0), axis=1)).mean()
```

```python
interactions
```

```python
calc_precision('prediction_user_based')
```

```python
from scipy.linalg import svd
U, sigma, V = svd(ratings)
```

```python
U.max()
```

```python
k = 100
s = np.diag(sigma[:k])
U = U[:, 0:k]
V = V[0:k, :]

round(s.sum(), 2)
```

```python
new_ratings = pd.DataFrame(
    U.dot(s).dot(V), index=ratings_m.index, columns=ratings_m.columns
)
top_k = 10
predictions = []

for personId in interactions.index:
    prediction = (
        new_ratings.loc[personId].sort_values(ascending=False).index.values
    )

    predictions.append(
        list(
            prediction[
                ~np.in1d(prediction, interactions.loc[personId, "true_train"])
            ]
        )[:top_k]
    )

interactions["prediction_svd"] = predictions

calc_precision("prediction_svd")
```
